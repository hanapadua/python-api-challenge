```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
from config import weather_api_key

# Import API key
#from weather_api_key import g_key

# create variable & csv path to cities
worldcities_to_load = "../../output_data/cities.csv"
worldcities = pd.read_csv(worldcities_to_load)
worldcities

# Generate cities amount
numberofcities = len(worldcities['City'].unique())
numberofcities

# import weather api key
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "metric"

# Build partial query URL
query_url = f"{url}appid={weather_api_key}&units={units}&q="


# start on scatter plots
```

```python
citys = [worldcities['City'].unique()]
```

```python

# set up lists to hold reponse info
lat = []
temp = []


# Loop through the list of cities and perform a request for data on each
for city in citys:
    response = requests.get(query_url + city).json()
    lat.append(response['coord']['lat'])
    temp.append(response['main']['temp'])

print(f"The latitude information received is: {lat}")
print(f"The temperature information received is: {temp}")

```

```python
# create a data frame from cities, lat, and temp
weather_dict = {
    "city": citys,
    "lat": lat,
    "temp": temp
}
weather_data = pd.DataFrame(weather_dict)
weather_data.head()
```
